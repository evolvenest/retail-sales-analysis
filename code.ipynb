```python
import pandas as pd          # import pandas
from google.colab import files

uploaded = files.upload()
df = pd.read_csv("data.csv") # read csv file
df.head()                   # show first 5 rows
```

```python
df.isnull().sum()   # check missing values
```

```python
df = df.dropna()   # remove missing values
```

```python
region_sales = df.groupby("Region")["Sales"].sum()  # total sales per region
region_sales
```

```python
region_sales.plot(kind="bar")   # bar chart
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], format='mixed')  # convert to date
df["Month"] = df["Order Date"].dt.month              # extract month

monthly_sales = df.groupby("Month")["Sales"].sum()   # sales per month
monthly_sales
```

```python
monthly_sales.plot(kind="line")   # line chart for trend
```

```python
category_sales = df.groupby("Category")["Sales"].sum()  # sales per category
category_sales
```

```python
category_sales.plot(kind="bar")  #bar chart for category
```

```python
import matplotlib.pyplot as plt
region_sales.plot(kind="bar")      # create chart
plt.title("Sales by Region")       # title
plt.savefig("region_sales.png")    # save image
plt.show()
```

```python
monthly_sales.plot(kind="line")
plt.title("Monthly Sales Trend")
plt.savefig("monthly_sales.png")
plt.show()
```

```python
category_sales.plot(kind="bar")
plt.title("Sales by Category")
plt.savefig("category_sales.png")
plt.show()
```
